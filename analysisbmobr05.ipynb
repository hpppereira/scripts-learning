# Data analysis from BMOBR05

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

%matplotlib inline
```

```python
'''
Load data
'''

pathname = os.environ['HOME'] + '/Dropbox/projetos/BMOP/Sistemas-BMOP/Processamento/dados/BMOBR05_CF1/op/'

filename = 'Dados_BMOBR05.csv'

#date of the accident
dac = pd.Timestamp('2016-04-21 11:00:00')

dd = pd.read_csv(pathname + filename, parse_dates=['date'], index_col=['date'])

#datetime index of the accident day
dac = np.where(dd.index == pd.Timestamp('2016-04-21 11:00:00'))[0][0]
```

```python
fig = plt.figure(figsize=(15,10))

ax1 = fig.add_subplot(311)
ax1.plot(dd.index, dd.hs,'b-')
ax1.plot([dd.index[dac], dd.index[dac]], [0, 5],'k--')
ax1.grid()
ax1.set_ylim(0,5)
ax1.set_xlim(dd.index[dac-200], dd.index[dac+200])

ax2 = fig.add_subplot(312)
ax2.plot(dd.index, dd.tp,'b.')
ax2.plot([dd.index[dac], dd.index[dac]], [0, 20],'k--')
ax2.grid()
ax2.set_ylim(0,20)
ax2.set_xlim(dd.index[dac-200], dd.index[dac+200])

ax3 = fig.add_subplot(313)
ax3.plot(dd.index, dd.dp,'b.')
ax3.plot([dd.index[dac], dd.index[dac]], [0, 360],'k--')
ax3.grid()
ax3.set_ylim(0,360)
ax3.set_xlim(dd.index[dac-200], dd.index[dac+200])

fig.savefig('fig/BMOBR05.png')
```
