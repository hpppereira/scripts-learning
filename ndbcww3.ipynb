```python
'''Comparinson between NDBC buoy and WW3 wave model'''
%matplotlib inline
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pathname = os.environ['HOME'] + '/Dropbox/database/'
ndbc = pd.read_csv(pathname + 'ndbc/realtime/51202.csv', sep=',', parse_dates=['date'], index_col=['date'])
nww3 = pd.read_csv(pathname + 'nww3/forecast/NWW3_NDBC_51202.csv', sep=',', parse_dates=['date'], index_col=['date'])
```

```python
fig = plt.figure(figsize=(14,12))
ax1 = fig.add_subplot(311)
ax1.plot(ndbc.index, ndbc.WVHT, '.b', nww3.index, nww3.hs, '.r')
ax1.set_xlim(ndbc.index[0], nww3.index[-1])
ax1.set_title('51202')
ax1.grid()
ax1.set_ylabel('Hs (m)')
ax2 = fig.add_subplot(312)
ax2.plot(ndbc.index, ndbc.DPD, '.b', nww3.index, nww3.tp, '.r')
ax2.set_xlim(ndbc.index[0], nww3.index[-1])
ax2.grid()
ax2.set_ylabel('Tp (s)')
ax3 = fig.add_subplot(313)
ax3.plot(ndbc.index, ndbc.MWD, '.b', nww3.index, nww3.dp, '.r')
ax3.set_ylim(0,360)
ax3.grid()
ax3.set_ylabel('Dp (deg)')
ax3.set_xlim(ndbc.index[0], nww3.index[-1])
```
