# Create an example text file and convert do JSON
- Used for ATM-Play

```python
# import libraries

import os
import datetime
import numpy as np
import pandas as pd
import matplotlib.pylab as pl
```

```python
# creating text file (dataframe)

todays_date = datetime.datetime.now().date()
index = pd.date_range(todays_date, periods=15*24, freq='H')


columns = ['ws',    # Wind Speed (m/s)
           'wd',    # Wind Direction (degrees)
           'wg',    # Wind Gusts (m/s)
           'ate',    # Air Temp (oC)
           'cld',    # Clouds (%)
           'prc',    # Precipitation (mm/h)
           'hs',    # Swell size (m)
           'tp',    # Swell Period (s)
           'dp',    # Swell Direction (degrees)
           'tid',    # Tides (m)
           'css',    # Surface currents speed (m/s)
           'csd',    # Surface currents direction (degrees)
           'wte',    # Water Temp (oC)
           'trb',    # Turbidity (TSS)
           'cla']    # Chlorophyll (mg/m3)

data = np.random.random_sample((len(index), len(columns)))

df = pd.DataFrame(data, index=index, columns=columns)

df.to_csv('JSON_csv_ATM_Play.csv', header=True, index=True, index_label='date')

df.to_json('JSON_ATM_Play.json', orient='columns', date_format='epoch',
            double_precision=2, force_ascii=True, date_unit='ms',
            default_handler=None, lines=False)



# df = pd.DataFrame(data, index=index, columns=columns)
```
